```python
import numpy as np

data = [
    47.2395401,
    7.261007309,
    44.05160141,
    51.41266251,
    53.76418304,
    45.81990051,
    50.70568466,
    48.59877777,
    54.5867691,
    34.69827271,
]

# Step 1: Calculate the median of the dataset
median = np.median(data)

# Step 2: Compute the absolute deviations from the median for each data point
absolute_deviations = [abs(x - median) for x in data]

# Step 3: Find the median of these absolute deviations
mad = np.median(absolute_deviations)

# Step 4: Find the anomaly index of all data points
anomaly_index = [x/mad for x in absolute_deviations]

# Step 5": Standardize anomaly index
std_anomaly_index = [x/1.4826 for x in anomaly_index]

print("Anomaly Index:", std_anomaly_index)
```
